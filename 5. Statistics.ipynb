```python
feature_list=['Age', 'Experience', 'Income', 'Family', 'CCAvg', 'Education','Mortgage', 'CreditCard', 
              'Securities Account', 'CD Account','Online']
def draw_distribution(df2,feature):
    """
    density
    """
    sns.distplot(df2[feature])
    plt.show()
```

```python
for i in feature_list:
    draw_distribution(df2,i)
    plt.show()
```
